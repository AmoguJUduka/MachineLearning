```python
# Import necessary libraries
import numpy as np
from keras.applications import vgg16
from keras.preprocessing import image
```

```python
# Build the model
model = vgg16.VGG16(weights='imagenet')
```

```python
# Mount your desired directory for image upload
from google.colab import drive
drive.mount('/content/drive')
```

```python
file_path = '/content/drive/MyDrive/sample-train/dogs/dog.977.jpg'
```

```python
import os

# Check if the path is correct
print(os.listdir('/content/drive/MyDrive/sample-train/dogs'))

# Load the file (for example, using OpenCV or PIL)
from PIL import Image

img = Image.open('/content/drive/MyDrive/sample-train/dogs/dog.977.jpg')
img.show()
```

```python
#Inspect the uploaded image
img = image.load_img(file_path, target_size=(224, 224))
img
```

```python
# Convert to Numpy
arr = image.img_to_array(img)
arr.shape
```

```python
# Expand Dimension
arr = np.expand_dims(arr, axis=0)
arr.shape
```

```python
# Preprocessing
arr = vgg16.preprocess_input(arr)
```

```python
# Predict
preds = model.predict(arr)
```

```python
# Predict for top 5
vgg16.decode_predictions(preds, top=5)
```

From VGG16, we can see that it classifies our image to be the German Shepherd dog breed because it has the highest probability when compared to other breeds
